```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the raw CSV file on GitHub
csv_url = 'https://raw.githubusercontent.com/supersweetmango/top10tardaytest1/main/TestReport1.csv'

# Read the CSV file from the GitHub repository
df = pd.read_csv(csv_url)

# Sort the DataFrame by 'Tardiness (Hours)' column in descending order
df_sorted = df.sort_values('Tardiness (Hours)', ascending=False)

# Select the top 10 rows
top_10 = df_sorted.head(10)

# Plot the bar chart using the modified DataFrame
top_10.plot.bar(x='Employee Number', y='Tardiness (Hours)')
plt.xlabel('Employee Number')
plt.ylabel('Tardiness (Hours)')
plt.title('Top 10 Employees - Tardiness by Employee')
plt.show()
```
